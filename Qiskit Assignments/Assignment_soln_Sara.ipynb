## Question - 2(a)

Quantum entanglement is a unique phenomena where two particles which are many light years far away from each other are connected in such a way that, if measurement is done on one particle, the other particle also gets the effect of it, and collapses to one of the state. Those two particle act as they're in the same wave function orin the overlapping wave function, but in reality they're from two different wave function, That's the main concept of EPR paradox and also a unique phenomena of quantum mechanics which can be used for the energy teleportation.

4 Bell States are given below:

1. **|Φ+> = (|00> + |11>)/√2**

2. **|Φ-> = (|00> - |11>)/√2**

3. **|Ψ+> = (|01> + |10>)/√2**

4. **|Ψ-> = (|01> - |10>)/√2**



#### **|Φ+>**

````python
#for 1 no bell state
bell_1 = QuantumCircuit(2)
bell_1.h(0)         #applying hadamard gate
bell_1.cx(0, 1)     #applying CNOT gate
%matplotlib inline
bell_1.draw('mpl', initial_state=True)
````

To create the **|ϕ+>** bell state, two qubits (initially as **|00>** state) is chosen, and a **Hadamard gate** is applied to one of the qubit to make the superposition of state:  

`1/√2( |00> + |10>)`

Then a **CNOT gate** is applied to both qubit (if Control qubit 0, then will remain the same; and if control qubit 1, then will flip the target qubit); so the output then becomes:

`1/√2( CNOT|00> + CNOT|10>)`

`1/√2( |00> + |11>)`; which is basically the |ϕ+> bell state

#### **|Φ->**

````python
#for 2 no bell state
bell_2 = QuantumCircuit(2)
bell_2.h(0)
bell_2.cx(0, 1)
bell_2.z(0)        #applying Z gate
%matplotlib inline
bell_2.draw('mpl', initial_state=True)
````

Here to Create the **|ϕ->** bell state, first have created the |ϕ+> state and then applied the unitary operator (**Z Gate**) on the 1st qubit, so the output becomes:


```
Z (1/√2( |00> + |11>)) =>  1/√2( |00> + ( -|1> ⊗ |1>)) => 1/√2( |00> - |11>)
```
Here Z gate is applied on the 1st qubit so when found |0> state it won't change, but if found |1> state, then will change to |-1>




#### **|Ψ+>**

````python

#for 3 no bell state
bell_3 = QuantumCircuit(2)
bell_3.h(0)
bell_3.cx(0, 1)
bell_3.x(0)        #applying X gate
%matplotlib inline
bell_3.draw('mpl', initial_state=True)
````

Here to Create the **|Ψ+>** bell state, first have created the |ϕ+> state and then applied the unitary operator (**X Gate**) on the 1st qubit, so the output becomes:


```
X (1/√2( |00> + |11>)) => 1/√2( |10> + |01>) => 1/√2( |01> + |10>)
```
Here X gate is applied on the 1st qubit so when found |0> state it will change to |1> state, and if found |1> state, then will change to |0>


#### **|Ψ->**

````python
#for 4 no bell state
bell_4 = QuantumCircuit(2)
bell_4.h(0)
bell_4.cx(0, 1)
bell_4.x(0)        #applying X gate
bell_4.z(0)        #applying Z gate
%matplotlib inline
bell_4.draw('mpl', initial_state=True)
````

Here to Create the **|Ψ->** bell state, first have created the |Ψ+> state and then applied the unitary operator (**Z Gate**) on the 1st qubit, so the output becomes:


```
Z (1/√2( |01> + |10>)) => 1/√2(|01>+(-|1>⊗|0>)) => 1/√2(|01>-|10>)
```
Here Z gate is applied on the 1st qubit so when found |0> state it will change to |1> state, and if found |1> state, then will change to |0>

## Question 3 (b)

 an Ansatz(parameterized quantum circuit)

````python
from qiskit.circuit import Parameter
n_qbits=3
theta_0= Parameter('00')   #taking parameter -> theta
theta_1= Parameter('01')
theta_2= Parameter('10')

qc= QuantumCircuit(n_qbits)

qc.h(0)  #appllying H to make superposition
qc.h(1)
qc.h(2)

qc.ry(theta_0,0)   #applying Rotation along Y axis with theta
qc.ry(theta_1,1)
qc.ry(theta_2,2)

qc.cx(0,1)   #applying cnot for entanglement
qc.cx(1,2)

qc.measure_all()
print(qc)
````

**Understanding about the circuit:**

an ansatz circuit is a circuit of quantum gates that are controlled by parameters,which can be adjusted to optimize the performance of the classification task or the QNN, Here the circuit was created with a Hadamard gate to make the superposition of states of qubits (which are initially at |000> state), then have applied the parameterized rotation (Ry gates) to adjust the state of each qbits, then CNOT is used to create entanglement between them. This Ansatz is used in QNN to enable learning by adjusting parameters during the training process.

## Question 4(a)

To make the state **Ψ = (|011⟩ - |100⟩)/√2**,

1st applying H gate to qubit-0 in a 3 qubit system, so the output becomes:
```
H(|000>) => (|000⟩ + |100⟩)/√2        [creating superposition]
```
Then applying X-gate to qubit-1, so the output becomes:
```
X((|000⟩ +|100⟩)/√2) => |010⟩+ |110⟩)/√2      [fliping the qubit-1]
```
Then applying CNOT gate to qubit-0 as control, and qubit-1 as target, so:


```
CNOT((|010⟩ +|110⟩)/√2) => (|010⟩ +|100⟩)/√2     [if qubit-0 is 1, then flipping the qubit-1]
```
Then again applying CNOT gate to qubit-1 as control, and qubit-2 as target, so:


```
CNOT((|010⟩+|100⟩)/√2) => (|011⟩ +|100⟩)/√2      [if qubit-1 is 1, then flipping the qubit-2]
```
Then applying Z gate on the qubit-0, so it becomes:


```
Z (|011⟩ +|100⟩)/√2 => (|011⟩ - |100⟩)/√2 ; which is the expected state
```

````python
qc = QuantumCircuit(3)   #creating a 3-qubit system
qc.h(0)     #applying H gate to make entagled state
qc.x(1)     #X gate: if 0, then 1; if 1 then 0
qc.cx(0,1)  #if control 0, then target no change; if control 1, then will flip the target
qc.cx(1,2)
qc.z(0)     #if 0, then no change; if 1 then will make it -1


%matplotlib inline
qc.draw('mpl')
````

We know,
$$
\begin{aligned}
& H|0\rangle=\left(\begin{array}{cc}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{array}\right)\left(\begin{array}{l}
1 \\
0
\end{array}\right)=\left(\begin{array}{c}
\frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}}
\end{array}\right)=|+\rangle \\
& H|1\rangle=\left(\begin{array}{cc}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{array}\right)\left(\begin{array}{l}
0 \\
1
\end{array}\right)=\left(\begin{array}{c}
\frac{1}{\sqrt{2}} \\
-\frac{1}{\sqrt{2}}
\end{array}\right)=|-\rangle \\
& H|+\rangle=\left(\begin{array}{cc}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{array}\right)\left(\begin{array}{l}
\frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}}
\end{array}\right)=\left(\begin{array}{l}
1 \\
0
\end{array}\right)=|0\rangle \\
& H|-\rangle=\left(\begin{array}{cc}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{array}\right)\left(\begin{array}{c}
\frac{1}{\sqrt{2}} \\
-\frac{1}{\sqrt{2}}
\end{array}\right)=\left(\begin{array}{l}
0 \\
1
\end{array}\right)=|1\rangle
\end{aligned}
$$
